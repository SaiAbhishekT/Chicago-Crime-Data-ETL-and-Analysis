### 3- Construct a query to retrieve the count of case incidents, count and percent (rounded to 3 decimal places) of arrests for each year;  order descending by arrested percent.
<span style="font-size:7px"><b>Hint:</b> 22 rows affected -- data[2015=26.451%] -- parse dates, uses "case when"</span>

```python
%%sql
WITH temp1 AS (
SELECT extract(year from cast(incident_date as timestamp)) as year, 100*sum(case when arrest = True then 1 else 0 end) arrested,
        100*cast(sum(case when arrest = True then 1 else 0 end) as float)/count(*) arrest_percent
FROM cc_cases 
GROUP BY year
)
select year,arrested,round(cast(arrest_percent as numeric),3) arr_percent from temp1;
```

```python
with engine.connect() as connection:
    df3 = pd.read_sql_query('''
    SELECT  cast (a.year as int) as year,
            concat(cast((a.arst_cnt/b.case_cnt)*100 as decimal(38,3)),'%%') as arrested_percent
    FROM    (
            SELECT  extract(year from incident_date) as year,
                    cast(count(1) as real) as arst_cnt
            FROM ycnghb.cc_cases
            WHERE arrest = TRUE
            group by year
            )a
    JOIN    (
            SELECT  extract(year from incident_date) as year,
                    cast(count(1) as real) as case_cnt
            FROM ycnghb.cc_cases
            group by year
            )b on a.year = b.year
    ORDER BY arrested_percent DESC
    ''', connection)
df3
```

### 5- For each year, which month is ranked #1 as having the greatest number of crime incidents with a primary crime description containing the term "NARCOTICS" or secondary crime desciption containing the phrase "GUN"? Display the year, month of the cases incident date, the primary and secondary crime description, the count of incidents and the ranking number. Order by year descending.

<span style="font-size:7px"><b>Hint:</b>22 rows affected -- data[2012-2, incidents=1942] ranking# should all be 1 -- involves parsing dates, a nested query with a window function and groups</span>

```python
%%sql
SELECT Year||'-'||Month as Year_Month,
        Num_of_incidents,
        foo.iucr_primary_desc,
        foo.iucr_secondary_desc
from
(SELECT extract( year from cast(t1.incident_date as timestamp)) as Year,
        extract(month from cast(t1.incident_date as timestamp)) as Month,
        count(*) as Num_of_incidents, t2.iucr_primary_desc, t3.iucr_secondary_desc,
        rank() over (partition by extract ( year from cast(incident_date as timestamp)) order by count(*) desc) as ranking
from cc_cases as t1
JOIN cc_iucr_codes_primary_descriptions as t2
ON t1.iucr_code = t2.iucr_code
JOIN cc_iucr_codes_secondary_descriptions as t3
ON t1.iucr_code = t3.iucr_code
WHERE (t2.iucr_primary_desc LIKE '%NARCOTICS%' or t3.iucr_secondary_desc LIKE '%GUN%' ) 
GROUP BY (Year,Month,t2.iucr_primary_desc,t3.iucr_secondary_desc)) as foo
WHERE ranking=1
ORDER BY Year DESC;
```

### 6 What is the average difference (in days, expressed as an integer) between the updated and incident dates for case arrests and primary and secondary crime descriptions ordered by the average difference in days.

<span style="font-size:7px"><b>Hint:</b>HINT: 538 rows affected -- output first row data[NARCOTICS,SALE/DEL DRUG PARAPHERNALIA,False,6] -- Involves Aggregate groups, parse date, calculate date difference</span>

```python
%%sql
select extract(year from cast(c.incident_date as timestamp)) as year,extract(month from cast(c.incident_date as timestamp)) as month
,sum(case when (pd.iucr_primary_desc like '%NARCOTICS%' or sd.iucr_secondary_desc like '%GUN%') then 1 else 0 end) 
from cc_iucr_codes_primary_descriptions pd join cc_cases c
using(iucr_code) join
cc_iucr_codes_secondary_descriptions sd using(iucr_code)
group by month,year
order by sum desc
;
```
